```python
import numpy as np
from time import time
from tqdm import trange
```

```python
n = 1000
A = np.random.randint(0, 10, (n, n)).astype(float)  # Matrix of shape nxn
A.shape
```

```python
B = np.random.randint(0, 10, n).astype(float)     # Vector of shape nx1
B.shape
```

```python
start_time1 = time()
for i in trange(1000):
    a = np.linalg.solve(A, B)
end_time1 = time()

print("Time for solve: ", end_time1 - start_time1)

start_time2 = time()
for i in trange(1000):
    a = np.linalg.inv(A).dot(B)
end_time2 = time()

print("Time for inv: ", end_time2 - start_time2)
```

Problem: solve for x: Ax=B, where A is nxn matrix, B is nx1 vector, and x is nx1 vector.

We can observe from above that np.linalg.inv method takes far more time than np.linalg.solve method for large matrices or if the calculations are repeated many times. The np.linalg.solve method does not calculate the inverse of $\text{A}$, it uses LU decomposition which factorizes $\text{A}$ then solves for $\text{x}$ using forward and backward substitution. Whereas, np.linalg.inv method calculates the inverse of the matrix $\text{A}$ using np.linalg.solve to solve $\text{AA}^{-1}=\text{I}$, where $\text{I}$ is the Identity matrix. Solving $\text{AA}^{-1}=\text{I}$ takes more operations as well as time than solving $\text{Ax}=\text{B}$ directly because here x is a vector of shape $(\text n,1)$ whereas $\text{A}^{-1}$ is a matrix of shape $(\text{n,n})$. Further, after calculating $\text{A}^{-1}$, we need to multiply it with $\text{B}$ to get the solution vector $\text{x}$, which again takes more time and takes more operations. Hence, np.linalg.solve method is preferred over np.linalg.inv method for solving linear equations. We also believe that lesser the number of floating point operations, lesser the error in the solution. Hence, np.linalg.solve method is likely to be more accurate than np.linalg.inv method but this is not a major concern as both methods are highly accurate.
