```python
import cv2
import matplotlib.pyplot as plt
import torch
import numpy as np
```

```python
# Read the image
image = cv2.imread('dog.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# Crop a 50x50 square from the image
crop1 = image[100:150, :50, :]

# Plot the cropped image
plt.imshow(crop1)
plt.axis('off')
plt.show()
```

```python
# Crop a 50x50 square from the image
crop2 = image[150:200, 150:200, :]

# Plot the cropped image
plt.imshow(crop2)
plt.axis('off')
plt.show()
```

```python
# Crop a 50x50 square from the image
crop3 = image[250:300, 100:150, :]

# Plot the cropped image
plt.imshow(crop3)
plt.axis('off')
plt.show()
```

```python
def factorize(A, k, device=torch.device("cpu")):
    """Factorize the matrix D into A and B"""
    A = A.to(device)
    # Randomly initialize A and B
    W = torch.randn(A.shape[0], k, requires_grad=True, device=device)
    H = torch.randn(k, A.shape[1], requires_grad=True, device=device)
    # Optimizer
    optimizer = torch.optim.Adam([W, H], lr=0.01)
    mask = ~torch.isnan(A)
    # Train the model
    for i in range(1000):
        # Compute the loss
        diff_matrix = torch.mm(W, H) - A
        diff_vector = diff_matrix[mask]
        loss = torch.norm(diff_vector)

        # Zero the gradients
        optimizer.zero_grad()

        # Backpropagate
        loss.backward()

        # Update the parameters
        optimizer.step()

    return W, H, loss
```

```python
dict_loss_1 = {}
dict_size_1 = {}
```

```python
rank = 5
crop1_r = crop1[:, :, 0]
crop1_g = crop1[:, :, 1]
crop1_b = crop1[:, :, 2]

crop1_r = torch.tensor(crop1_r)
crop1_g = torch.tensor(crop1_g)
crop1_b = torch.tensor(crop1_b)

W1_r, H1_r, loss1_r = factorize(crop1_r, rank)
W1_g, H1_g, loss1_g = factorize(crop1_g, rank)
W1_b, H1_b, loss1_b = factorize(crop1_b, rank)

result1 = torch.zeros(crop1.shape)
result1[:, :, 0] = torch.mm(W1_r, H1_r)
result1[:, :, 1] = torch.mm(W1_g, H1_g)
result1[:, :, 2] = torch.mm(W1_b, H1_b)

result1 = result1.detach().numpy()
result1 = result1.astype(np.uint8)

dict_loss_1[rank] = np.sqrt(loss1_r.detach().numpy()**2 + loss1_g.detach().numpy()**2 + loss1_b.detach().numpy()**2)
dict_size_1[rank] = (W1_b.nbytes + H1_b.nbytes)*3

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(crop1)
ax[0].set_title('Original')
ax[0].axis('off')
ax[1].imshow(result1)
ax[1].set_title('Reconstructed')
ax[1].axis('off')
plt.title(f"Image Reconstruction for rank = {rank}")
plt.show()
```
